# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
import pandas as pd 
bestseller = pd.read_csv("../data/bestsellers.csv")
```

```python
#Find the books written by Pete Souza
bestseller[bestseller["Author"]=="Pete Souza"]
```

```python
# Find the books that are under 10 dollars
bestseller[bestseller["Price"]<10]
```

```python
#Find the books that have a price between 50 and 60 dollars
bestseller[bestseller["Price"].between(50,60)]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens
bestseller[(bestseller["Author"]=="Kristin Hannah")| (bestseller["Author"]=="Andy Weir")|(bestseller["Author"]=="Delia Owens")]
```

```python
#Find the Non Fiction books that are rated 4.9
bestseller[(bestseller["Genre"]=="Non Fiction") & (bestseller["User Rating"]==4.9)]
```

```python
#Find the fiction book with the lowest User Rating
bestseller[bestseller["Genre"]=="Fiction"].sort_values("User Rating").iloc[0]
```

```python
#Find 2012's top 5 Fiction books with the most Reviews
bestseller[(bestseller["Genre"]=="Fiction") & (bestseller["Year"]==2012)].sort_values("User Rating", ascending=False).head(5)
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5
bestseller[bestseller["User Rating"]< 4.5].Author.value_counts().head(5).plot(kind= 'bar')
```
